```python
import os
os.add_dll_directory(r'C:\Users\curti\OpenSlide\bin')

import openslide

file_path = r'C:\Users\curti\PycharmProjects\Crary_Lab\slides\41998.svs'

print(f"File exists: {os.path.exists(file_path)}")
print(f"File size: {os.path.getsize(file_path)} bytes")
print(f"Absolute path: {os.path.abspath(file_path)}")

try:
    slide = openslide.OpenSlide(file_path)
    print("Success!")
    print(f"Dimensions: {slide.dimensions}")
    print(f"Vendor: {slide.properties.get('openslide.vendor', 'Unknown')}")
except openslide.OpenSlideError as e:
    print(f"OpenSlide error: {e}")
except Exception as e:
    print(f"Other error: {e}")
```
